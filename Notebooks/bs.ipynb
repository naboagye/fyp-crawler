```python
import requests
from bs4 import BeautifulSoup

URL = 'https://www.gov.uk/guidance/transport-measures-to-protect-the-uk-from-variant-strains-of-covid-19#red-list-travel-ban-countries'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.text)
```

```python
results = soup.find(id='gem-c-govspeak')
```

```python
print(results.prettify())
```

```python
job_elems = soup.find('div', class_='govspeak')
```

```python
print(job_elems)
```

```python
countries = job_elems.find_next('ul').find_all('li')
```

```python
countries
```

```python
for country in countries:
    print(country.text)
```
